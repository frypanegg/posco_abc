```python
class NumberPrinter:
    def print_numbers(self, count):
        for i in range(count):
            print(i)

printer = NumberPrinter()
printer.print_numbers(5)
```

```python
def classify_age(age):
    if age < 0:
        return "Invalid age"
    if age < 13:
        return "Child"
    if age < 20:
        return "Teenager"
    if age < 65:
        return "Adult"
    return "Senior"

print(classify_age(25))
```
